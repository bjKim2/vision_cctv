```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris

iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.DataFrame(iris.target, columns=['target'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100,10), solver='lbfgs', alpha=100, random_state=1)  # 히든레이어와 알파값을 조정하며 분류, 알파값이 작을수록 복잡도 상승
clf.fit(X_train,y_train)
```

```python
pred = clf.predict(X_test)
pred_proba = clf.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

# SVM  - 마진을 최대로 하는 초평면을 결정하는 모델, C 로 에러 허용

# SVM with Kernel - 선형관계가 아닌경우 사용

- 장점 :예측의 정확도가 높음
- 단점 : C를 결정해야함, 모형 구축에 시간이 오래걸림

```python
from sklearn import svm
```

```python
iris=load_iris()
X=iris.data[:,:2]
y=iris.target

# C가 클수록 error를 적게 허용하는것
# C가 무한대이면, 오분류가 하나도 없는 SVM상황이 될 것
C=1
clf=svm.SVC(kernel='linear',C=C) # kernel='linear'는 커널을 안주는 모형
clf.fit(X,y)
```

## kernel SVM 적합 및 비교

```python
# LinearSVC - 위 모형과 비슷하지만 결과가 다르게 나옴 -> 사용한 알고리즘이 달라서, loss funtion이 달라서
clf=svm.LinearSVC(C=C,max_iter=10000)
clf.fit(X,y)
y_pred=clf.predict(X)
confusion_matrix(y,y_pred)
```

```python
# radial basis function
clf=svm.SVC(kernel='rbf',gamma=0.7,C=C,max_iter=10000)
clf.fit(X,y)
y_pred=clf.predict(X)
confusion_matrix(y,y_pred)
```

```python
# polynomial kernel
clf=svm.SVC(kernel='poly',degree=3,C=C,gamma='auto')
clf.fit(X,y)
y_pred=clf.predict(X)
confusion_matrix(y,y_pred)
```

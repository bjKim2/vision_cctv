```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import random

```

```python
x = np.array(random.sample(range(1,1000),k=150))
x = x.reshape(-1,1)
y = x **2 + 5  
```

```python
print(x.shape,y.shape)
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(4, input_shape = (1,),activation='relu'))
model.add(keras.layers.Dense(8,activation='relu'))
model.add(keras.layers.Dense(8,1))
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.8), loss=keras.losses.MSE)
```

```python
model.fit(x,y,epochs=1000)
```

```python
model.predict([[6]])
```

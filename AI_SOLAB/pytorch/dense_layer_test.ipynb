```python
import torch
import numpy as np
import random
from torch import nn
```

```python

x = torch.tensor(random.sample(range(1,1000),k=150))


x = x.reshape(-1,2).type(torch.float32)
y = x 
for i in range(len(x)):
    y[i] = x[i,0] * x[i,1] + 5

y = y.reshape(-1,1).type(torch.float32)

print(x.shape, y.shape)
print(x[1],y[1])
print(x.dtype,y.dtype)
```

```python
2**2 +5
```

```python
class dataset():
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __len__(self):
        return len(self.x)
    def __getitem__(self,idx):

        return x[idx],y[idx]
```

```python
dl = torch.utils.data.DataLoader(
    dataset = dataset(x,y),
    batch_size = 64,
    shuffle = True,
    drop_last = True

)
```

```python
class model(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Sequential(
            nn.Linear(2,256 ,bias=False),
            nn.Linear(256,512),
            nn.Linear(512,1)            
            )
    def forward(self,x):
        out = self.layer1(x)
        return out
model_a = model().to(device='cuda')
```

```python
criterion = torch.nn.MSELoss(reduction='mean')
optimizer = torch.optim.Adam(model_a.parameters(),lr= 0.01)
```

```python

total_batch = len(dl)
for epoch in range(100000):
    avg_cost = 0

    for a,b in dl:
        a = a.type(torch.float32).to('cuda')
        b = b.type(torch.float32).to('cuda')

        optimizer.zero_grad()
        pred = model_a(a)
        loss = criterion(pred, b)
        loss.backward()

        optimizer.step()

        avg_cost += loss/total_batch

    print('[Epoch: {:>4}] cost = {:>.9}'.format(epoch + 1 , avg_cost))
```

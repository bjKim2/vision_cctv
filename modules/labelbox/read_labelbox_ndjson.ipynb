```python
import json

with open('./export_json_file/tank_video_1.ndjson' , 'r') as f:
    datas = []
    for line in f:
        data = json.loads(line)
        datas.append(data)
```

```python
def tlhw2xywh(tlhw,w,h):
    return [(tlhw[1]  + tlhw[3]/2) / w , (tlhw[0] + tlhw[2]/2) / h , tlhw[3] /w ,tlhw[2] / h ]
```

```python
import cv2
import os 
import shutil
```

```python
os.mkdir('images')
os.mkdir('labels')
```

```python
count = 0

for i in datas:
    try:
        img_name = (i['data_row']['external_id'])
        img_file = '../../../labelbox-python/tank_images/' + img_name
        img = cv2.imread(img_file)
        h, w = img.shape[0], img.shape[1]
        bbox = i['projects']['clhzs6x9s000i07u91slk7dh6']['labels'][0]['annotations']['objects'][0]['bounding_box']
        bbox = tlhw2xywh(list(bbox.values()) ,w ,h )
        bbox = list(map(str,bbox))
        shutil.copy(img_file, 'images/' + img_name)
        with open('labels/' + img_name[:-4] + '.txt' , 'w') as f:
            f.write('0 ' +' '.join(bbox) + ' \n')
    except Exception as e:
        print(e)
        count+=1
print(count)
```

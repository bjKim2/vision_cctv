```python
import cv2
# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../../temp/yolov7/test_images/tank/download.jpeg')
# img = img[...,::-1]
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img.shape
```

```python
# img = img[...,::-1]
# img.shape
```

```python

# "top": 120.0, "left": 168.0, "height": 159.0, "width": 255.0
c1=(168, 140)
c2=(320, 279)
color=(255,0,0)
label = 'tank'
img = cv2.rectangle(img, pt1=c1, pt2 =c2, color=color, thickness = 1, lineType=cv2.LINE_AA)  # filled
img = cv2.putText(img, label + ' 0.59', (c1[0], c1[1] - 2), 0, 1/2, color, thickness=1, lineType=cv2.LINE_AA)
```

```python
c1=(290, 120)
c2=(423, 180)
color=(255,0,0)
label = 'tank'
img = cv2.rectangle(img, pt1=c1, pt2 =c2, color=color, thickness = 1, lineType=cv2.LINE_AA)  # filled
img = cv2.putText(img, label + ' 0.38', (c1[0], c1[1] - 2), 0, 1/2, color, thickness=1, lineType=cv2.LINE_AA)
```

```python
plt.imshow(img)
plt.show()
```

```python
def bbox_converter(bbox,x,y):
    return [(bbox[0]+bbox[2])/2/x , (bbox[1]+bbox[3])/2/y, (bbox[2] -bbox[0])/x, (bbox[3] - bbox[1])/y ]


def json2txt(json_path, name ,ratio):
    with open(json_path, 'r') as f:
        data = json.load(f)
    data_anno = data['annotations']
    frame_size = data['video']['total_frame']
    
    x = data['video']['resolution'][0]
    y = data['video']['resolution'][1]
    
    frame = []
    
    txt_data = {}
    for i in data_anno:
        
        if i['id'].split('_')[0] != 'person':
            continue
        if i['frame'] not in frame:
            frame.append(i['frame'])
            txt_data[i['frame']] = [bbox_converter(i['bbox'],x,y)]
        else:
            txt_data[i['frame']].append(bbox_converter(i['bbox'],x,y))

    select_num = int(frame_size * ratio)
    if select_num > len(frame):
        select_num = len(frame)
    random.seed(42)
    random.shuffle(frame)
    
    for i in range(select_num):
        
        with open("./lbls3/%s_%d.txt"%(name,frame[i]), 'w') as f:
            # if i in frame:
            for j in txt_data[frame[i]]:
                f.write('0 %f %f %f %f\n'%(j[0],j[1],j[2],j[3]))   
        print("./lbls3/%s_%d.txt"%(name,frame[i]))
    return frame[:select_num]
        
```

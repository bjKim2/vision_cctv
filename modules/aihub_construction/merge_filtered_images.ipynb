```python
from glob import glob

imgs = glob('/media/tekim/tekim/tmp1/completed_labeling_files/*/images/*')
filtered_imgs = glob('/media/tekim/tekim/tmp1/completed_labeling_files/*/filtered_images/*')
labels = glob('/media/tekim/tekim/tmp1/completed_labeling_files/*/labels/*')
filtered_labels = glob('/media/tekim/tekim/tmp1/completed_labeling_files/*/filtered_labels/*')
```

```python
import os
import shutil
```

```python
dir_path = '/media/tekim/tekim/datasets/aihub_safety/human_car_ver2/filtering/images'
for i in filtered_imgs:
    try:
        # shutil.move( os.path.dirname(dir_path) + '/filtered_images/'  + os.path.basename(i), os.path.dirname(i) )
        shutil.move(i, os.path.dirname(dir_path) + '/filtered_images' )
    except:
        print(1)
for i in imgs:
    try:
        # shutil.move( os.path.dirname(dir_path) + '/images/'  + os.path.basename(i), os.path.dirname(i) )
        shutil.move(i, os.path.dirname(dir_path) + '/images' )
    except:
        print(1)
```

```python
dir_path = '/media/tekim/tekim/datasets/aihub_safety/human_car_ver2/filtering/labels'
for i in filtered_labels:
    try:
        shutil.move(i, os.path.dirname(dir_path) + '/filtered_labels' )
    except:
        print(1)
for i in labels:
    try:
        shutil.move(i,dir_path)
    except:
        print(1)
```

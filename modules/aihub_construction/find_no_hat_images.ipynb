```python
from glob import glob
import cv2
from tqdm import tqdm
import os
import shutil

labels = glob('../../../datasets/aihub_safety/human_car_ver2/labels/*')

imgs = []
for i in tqdm(labels):
    with open(i, 'r') as f:
        datas = f.readlines()
    
    cls_five = False

    for j in datas:
        anno = j.strip()
        cls = int(anno.split()[0])

        if cls == 7:
            cls_five = True
    
    if cls_five:
        img = i.replace('labels','images')[:-3] + 'jpg'
        imgs.append(img)        
```

```python

for i in tqdm(imgs):
    shutil.copy(i,'../../../datasets/aihub_safety/human_car_ver2/cls5_images' )
    shutil.copy(i.replace('images','labels')[:-3] + 'txt','../../../datasets/aihub_safety/human_car_ver2/cls5_labels')
```

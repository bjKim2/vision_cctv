```python
# 복리-단리의 차이가 이율을 달리 했을 때 그래프

import matplotlib.pyplot as plt
import numpy as np

n = 10 # 년 수
baek = np.arange(100) * 0.01 + 1
```

```python
baek2 = baek.copy()
```

```python
baek2
```

```python
for _ in range(n):
    baek2 = np.multiply(baek2,baek)
```

```python
baek2
```

```python
baek2 -= ((baek-1) *n)
```

```python
baek2
```

```python
# 1 을 10년간 다른 이율로 할 때 복리 그래프

plt.plot(
    baek2
    
)
```
